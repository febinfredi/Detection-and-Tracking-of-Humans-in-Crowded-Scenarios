```python
from pathlib import Path

import numpy as np
import cv2
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
jpg_paths = list(Path('raw/Images').rglob('*.jpg'))
img_widths, img_heights = [], []
for i, jpg_path in enumerate(jpg_paths):
    if i % 1000 == 0 or i == len(jpg_paths) - 1:
        print('Processing images %d/%d' % (i, len(jpg_paths)))
    img = cv2.imread(jpg_path.as_posix())
    assert img is not None
    img_h, img_w, img_c = img.shape
    assert img_c == 3
    img_widths.append(img_w)
    img_heights.append(img_h)

img_widths = np.array(img_widths)
img_heights = np.array(img_heights)
```

```python
img_aspects = img_widths / img_heights

idx = img_widths.argmin()
print('image with min width  (%5d): %s' % (img_widths[idx], jpg_paths[idx].name))
idx = img_widths.argmax()
print('image with max width  (%5d): %s' % (img_widths[idx], jpg_paths[idx].name))
idx = img_heights.argmin()
print('image with min height (%5d): %s' % (img_heights[idx], jpg_paths[idx].name))
idx = img_heights.argmax()
print('image with max height (%5d): %s' % (img_heights[idx], jpg_paths[idx].name))
idx = img_aspects.argmin()
print('min aspect ratio    (%7.5f): %s' % (img_aspects[idx], jpg_paths[idx].name))
idx = img_aspects.argmax()
print('max aspect ratio    (%7.5f): %s' % (img_aspects[idx], jpg_paths[idx].name))
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(img_widths)
plt.title('image widths')
plt.subplot(1, 2, 2)
plt.hist(img_heights)
plt.title('image heights')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(1, 1, 1)
plt.scatter(img_widths, img_heights)
plt.title('image widths/heights')
plt.xlabel('width')
plt.ylabel('height')

plt.show()
```
